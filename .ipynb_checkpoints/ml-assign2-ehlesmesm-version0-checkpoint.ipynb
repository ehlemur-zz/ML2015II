# Preliminaries

```python
import pylab
import numpy as np
from IPython.display import display, Math, Latex

%matplotlib inline


def readInput(filename):
    data_lst = [[],[],[]]
    with open(filename) as f:
        for line in f:
            x, y, c = map(float, line.split(','))
            c = int(c)
            data_lst[c].append([x, y])
    classes = []
    for i in range(3):
        classes.append(np.array(data_lst[i]))
        np.random.shuffle(classes[-1])
    return classes

classes = readInput("data.txt")
lengths = np.array([float(len(cls))  for cls in classes])
```

# Analysis for classes 1 and 2

## Gaussian distribution for classes 1 and 2 assuming equal covariance

```python
means = np.array([np.mean(cls, axis=0)  for cls in classes])

cov1And2 = np.cov(np.append(classes[1] - means[1], classes[2] - means[2], axis=0).T)

print "Assuming equal covariance for classes 1 and 2 we have"
display(Math(r'C_1 \sim \mathcal{N}_2\left(\mathbf{\mu}_1,\mathbf{\Sigma}\right)'))
display(Math(r'C_2 \sim \mathcal{N}_2\left(\mathbf{\mu}_2,\mathbf{\Sigma}\right)'))

print "Where"

display(Math(r'\mathbf{\mu}_1='))
print means[1]

display(Math(r'\mathbf{\mu}_1='))
print means[2]

display(Math(r'\mathbf{\Sigma}='))
print cov1And2
```

## Discriminant functions for classes 1 and 2

Writing the discriminant function as $g_i(\mathbf{x})=\log p(\mathbf{x}|C_i)+log(C_i)$.

Assuming that $p(\mathbf{x}|C_i)\sim\mathcal{N}_2(\mathbf{\mu}_i, \mathbf{\Sigma})$ we obtain $g_i(\mathbf{x})=-\log 2\pi - \frac{1}{2}\log |\Sigma| - \frac{1}{2}(\mathbf{x}-\mathbf{\mu}_i)^T\Sigma^{-1}(\mathbf{x}-\mathbf{\mu}_i) + \log C_i$.
We can drop the term $-\log 2\pi - \frac{1}{2}\log |\Sigma|$ since it appears in both classes.

Expanding and regrouping we obtain $-\frac{1}{2}(
\mathbf{x}^T\Sigma^{-1}\mathbf{x}
- \mathbf{x}_i^T\Sigma^{-1}\mathbf{\mu}_i
- \mathbf{\mu}_i^T\Sigma^{-1}\mathbf{x}
+ \mathbf{\mu}_i^T\Sigma^{-1}\mathbf{\mu}_i)
+ \log C_i = 
-\frac{1}{2} \mathbf{x}^T\Sigma^{-1}\mathbf{x}
+\mathbf{\mu}_i^T\Sigma^{-1}\mathbf{x}
+ (-\frac{1}{2} \mathbf{\mu}_i^T\Sigma^{-1}\mathbf{\mu}_i
+ \log C_i)$. 
We can also drop the term $-\frac{1}{2} \mathbf{x}^T\Sigma^{-1}\mathbf{x}$, since is equal for all clases, and we obtain the equation for a line:
$g_i(\mathbf{x})=\mathbf{w}_i^T\mathbf{x}+c_i$, where $\mathbf{w}_i=\Sigma^{-1}\mu_i$ and $c_i = \frac{-1}{2}\mu_i^T\Sigma^{-1}\mu_i + \log C_i$.

```python
def getDiscriminantSameMean(C, mean, cov):
    covInv = np.linalg.inv(cov)
    w = np.dot(covInv, mean)
    c = -.5 * np.dot(np.dot(mean.T, covInv), mean) + np.log(C)
    
    return (lambda x: np.dot(w.T, x) + c), w, c
```

```python
_, w1, c1 = getDiscriminantSameMean(lengths[1] / (lengths[1] + lengths[2]), means[1], cov1And2)
print "For the first class, the discriminant function is given by"
display(Math(r'g_1(\mathbf{x})=\mathbf{w}_1^T\mathbf{x}+c_1'))
print "where"
display(Math(r'\mathbf{w}_1='))
print w1
display(Math(r'c_1='))
print c1
print

_, w2, c2 = getDiscriminantSameMean(lengths[2] / (lengths[1] + lengths[2]), means[2], cov1And2)
print "For the second class, the discriminant function is given by"
display(Math(r'g_2(\mathbf{x})=\mathbf{w}_2^T\mathbf{x}+c_2'))
print "where"
display(Math(r'\mathbf{w}_2='))
print w2
display(Math(r'c_2='))
print c2
```

## Regions and boundary for classes 1 and 2

Recalling that the region for class 2 is given by $\{\mathbf{x}\in\mathbb{R}^2 : g_2(x) > g_1(x)\}$ or equivalently, $\{\mathbf{x}\in\mathbb{R}^2 : g_2(x) - g_1(x) > 0\}$ we can solve $g_2(x) - g_1(x) = 0$ to find the boundary. From this we can see that the equation for the boundary, must be a line, since it has the form $(\mathbf{w}_2 - \mathbf{w}_1)^T + (c_2 - c_1) = 0$. 

In this particular case, we obtain that the equation for the line is given by $\mathbf{w}^T\mathbf{x}+c=0$ where:

```python
display(Math(r'\mathbf{w}='))
print (w2-w1)
display(Math(r'c='))
print (c2-c1)
```

Next we plot the region for both classes and the line between them.

```python
def plot_regions_and_line(C1, x_min=2, y_min=-1, x_max=14, y_max=9.3, delta=0.04):
    classes_label = ["C1 p=" + str(C1), "C2 p=" + str(1-C1)]
    classes_colors = ["r", "g"]
    
    g1 = getDiscriminantSameMean(C1, means[1], cov1And2)[0]
    g2 = getDiscriminantSameMean(1 - C1, means[2], cov1And2)[0]
    
    def classify(x):
        return int(g1(x) < g2(x))
    
    pylab.clf()
    
    extent = (x_min,x_max,y_min,y_max)
    
    x = np.arange(x_min, x_max, delta)
    y = np.arange(y_min, y_max, delta)
    X, Y = np.meshgrid(x, y)
    Z = np.array([[classify(np.array([xi, yi]))  for xi in x] for yi in y])
    
    levels = np.arange(-1, 1.001, 1)
    
    norm = pylab.cm.colors.Normalize(vmax=2, vmin=-1)
    cmap = pylab.cm.PRGn
    
    cset1 = pylab.contourf(X, Y, Z, levels,
                            cmap=pylab.cm.get_cmap(cmap,3),
                            norm=norm,
                            hold='on')
    
    cset2 = pylab.contour(X, Y, Z, cset1.levels,
                            colors = 'k',
                            hold='on')
    
    for c in cset2.collections:
        c.set_linestyle('solid')
        
    class1 = pylab.scatter(classes[1][:,0], classes[1][:,1], c=classes_colors[0])
    class2 = pylab.scatter(classes[2][:,0], classes[2][:,1], c=classes_colors[1])
    
    pylab.legend((class1, class2),
           (classes_label[0], classes_label[1]),
           scatterpoints=1,
           loc='lower left',
           ncol=1,
           fontsize=8)
```

```python
plot_regions_and_line(lengths[1] / (lengths[1] + lengths[2]))
```

If we change the prior probability of the classes, the boundary moves towards the mean of the class with least probability, as we can see in the following diagrams.

```python
for i in range(5):
    plot_regions_and_line(0.1 + i * 0.2)
    pylab.show()
```

# Analysis for three classes

## Probability distribution for each class

The mean has already been calculated, we now proceed to calculate the covariance matrix for each class.

```python
covs = np.array([np.cov(cls.T)  for cls in classes])

for i in range(3):
    cls = str(i)
    display(Math(r'C_' + cls + r'\sim \mathcal{N}_2(\mu_' + cls + r', \mathbf{\Sigma}_' + cls + ')'))
    print "where"
    display(Math(r'\mu_' + cls + r'='))
    print means[0]
    display(Math(r'\mathbf{\Sigma}_' + cls + r'='))
    print covs[0]
    print
```

## Discriminant function

Writing the discriminant function as $g_i(\mathbf{x})=\log p(\mathbf{x}|C_i)+log(C_i)$.

Assuming that $p(\mathbf{x}|C_i)\sim\mathcal{N}_2(\mathbf{\mu}_i, \mathbf{\Sigma}_i)$ we obtain $g_i(\mathbf{x})=-\log 2\pi - \frac{1}{2}\log |\Sigma_i| - \frac{1}{2}(\mathbf{x}-\mathbf{\mu}_i)^T\Sigma_i^{-1}(\mathbf{x}-\mathbf{\mu}_i) + \log C_i$.
We can drop the term $-\log 2\pi$ since it appears in all classes.

Expanding and regrouping we obtain 

$-\frac{1}{2}(
\mathbf{x}^T\Sigma_i^{-1}\mathbf{x}
- \mathbf{x}_i^T\Sigma^{-1}\mathbf{\mu}_i
- \mathbf{\mu}_i^T\Sigma^{-1}\mathbf{x}
+ \mathbf{\mu}_i^T\Sigma^{-1}\mathbf{\mu}_i)
+ \log C_i
- \frac{1}{2}\log |\Sigma_i|
= 
-\frac{1}{2} \mathbf{x}^T\Sigma^{-1}\mathbf{x}
+\mathbf{\mu}_i^T\Sigma^{-1}\mathbf{x}
+ (-\frac{1}{2} \mathbf{\mu}_i^T\Sigma^{-1}\mathbf{\mu}_i
+ \log C_i
- \frac{1}{2}\log |\Sigma_i|)$. 

We obtain a quadratic equation: 
$g_i(\mathbf{x})=\mathbf{x}^T\mathbf{W}_i\mathbf{x} + \mathbf{w}_i^T\mathbf{x}+c_i$, where $\mathbf{W}_i=\frac{-1}{2}\Sigma_i^{-1}$, $\mathbf{w}_i=\Sigma_i^{-1}\mu_i$ and $c_i = \frac{-1}{2}\mu_i^T\Sigma_i^{-1}\mu_i + \log C_i - \frac{1}{2}\log |\Sigma_i|$.

```python
def getDiscriminant(C, mean, cov):
    covInv = np.linalg.inv(cov)
    W = -.5 * covInv
    w = np.dot(covInv, mean)
    c = -.5 * (np.dot(np.dot(mean.T, covInv), mean) + np.log(np.linalg.det(cov))) + np.log(C)
    
    return (lambda x: np.dot(np.dot(x.T, W), x) + np.dot(w.T, x) + c), W, w, c
```

In this particular case, the discriminants are given by

```python
totalSum = sum(lengths)
ordinal = "zeroth", "first", "second"
for i in range(3):
    cls = str(i)
    _, W, w, c = getDiscriminant(lengths[1] / totalSum, means[i], covs[i])
    print "For the " + ordinal[i] + " class, the discriminant function is given by"
    display(Math(r'g_' + cls + r'(\mathbf{x})=\mathbf{x}^T\mathbf{W}' + cls + r'\mathbf{x}+\mathbf{w}_' + cls + r'^T\mathbf{x}+c_' + cls))
    print "where"
    display(Math(r'\mathbf{W}_' + cls + '='))
    print W
    display(Math(r'\mathbf{w}_' + cls + '='))
    print w
    display(Math(r'c_' + cls + '='))
    print c
    print
```

## Regions and boundaries for all classes

We now plot the regions and boundaries corresponding to each class

```python
def plot_regions_and_boundaries(x_min=0, y_min=-1, x_max=14, y_max=9.3, delta=0.04):
    C = [lengths[i] / totalSum  for i in range(3)]
    classesLabels = ["C" + str(i) + " p=" + str(C[i])  for i in range(3)]
    classesColors = ["y", "r", "g"]
    
    g = [getDiscriminant(C[i], means[i], covs[i])[0]  for i in range(3)]
    
    def classify(x):
        return np.argmax(np.array([g[i](x)  for i in range(3)]))
    
    pylab.clf()
    
    extent = (x_min,x_max,y_min,y_max)
    
    x = np.arange(x_min, x_max, delta)
    y = np.arange(y_min, y_max, delta)
    X, Y = np.meshgrid(x, y)
    Z = np.array([[classify(np.array([xi, yi]))  for xi in x] for yi in y])
    
    levels = np.arange(-1, 2.001, 1)
    
    norm = pylab.cm.colors.Normalize(vmax=3, vmin=-1)
    cmap = pylab.cm.PRGn
    
    cset1 = pylab.contourf(X, Y, Z, levels,
                            cmap=pylab.cm.get_cmap(cmap,3),
                            norm=norm,
                            hold='on')
    
    cset2 = pylab.contour(X, Y, Z, cset1.levels,
                            colors = 'k',
                            hold='on')
    
    for c in cset2.collections:
        c.set_linestyle('solid')

    scatterPlots = [pylab.scatter(classes[i][:,0], classes[i][:,1], c=classesColors[i])  for i in range(3)]
    
    pylab.legend(scatterPlots,
           classesLabels,
           scatterpoints=1,
           loc='lower left',
           ncol=1,
           fontsize=8)
```

# Evaluation as a classifier

First, we separate the training data from the test data and calculate the mean and variance for each of the classes in the training set.

```python
trainSize = lengths * .8
train, test = zip(*[np.split(classes[i], [trainSize[i]]) for i in range(3)])
train = np.array(train)
test = np.array(test)

trainMeans = np.array([np.mean(cls, axis=0)  for cls in train])
trainCovs = np.array([np.cov(cls.T)  for cls in train])
```

We define the $\mathtt{getLogNormal}$ function, which returns the function $log(p(x))$ where $p(x) \sim \mathcal{N}_2(\mathtt{mean}, \mathtt{cov})$.

```python
def getLogNormal(mean, cov):
    covInv = np.linalg.inv(cov)
    W = -.5 * covInv
    w = np.dot(covInv, mean)
    c = -.5 * (np.dot(np.dot(mean.T, covInv), mean) + np.log(np.linalg.det(cov)) + np.log(2 * np.pi))
     
    return lambda x: np.dot(np.dot(x.T, W), x) + np.dot(w.T, x) + c
```

We then use this function to calculate the functions $log(p(x|C_i))$ and we calculate $p(C_i)$ as well.

```python
totalTrainSum = sum(trainSize)
C = np.array([trainSize[i] / totalTrainSum  for i in range(3)])
logNormals = [getLogNormal(trainMeans[i], trainCovs[i])  for i in range(3)]
```

Finally, we define a function $\mathtt{classify}$ that calculates $log(p(C_i|x))$ as well as $log(1-\lambda)$ and chooses the class with the highest probability, or rejects if none is greater than $1-\lambda$, where $\lambda$ is the loss parameter.

```python
def classify(x, loss):
    condProbs = np.array([logNormals[i](x)  for i in range(3)])  # condProbs[i] = log(p(x|C_i))
    logP = np.log(np.sum(np.exp(condProbs) * C))                 # logP[i] = log(sum p(x|C_i) C_i) = log(p(x))
    condProbs -= logP - np.log(C)                                # condProbs[i] = log(p(x|C_i)) - log(p(x)) + log(C_i)
                                                                 #              = log(p(x|C_i) p(C_i) / p(x))
                                                                 #              = log(p(C_i|x))
    condProbs = np.append(condProbs, [np.log(1 - loss)])         # the loss function
    return np.argmax(condProbs)
```

We plot the regions and boundaries calculated for the train set with a loss value $\lambda$ of $0.1$, as well as the training points.

```python
def plot_regions_and_boundaries2(classify, loss, x_min=0, y_min=-1, x_max=14, y_max=9.3, delta=0.04):
    pylab.clf()
    
    extent = (x_min,x_max,y_min,y_max)
    
    x = np.arange(x_min, x_max, delta)
    y = np.arange(y_min, y_max, delta)
    X, Y = np.meshgrid(x, y)
    Z = np.array([[classify(np.array([xi, yi]), loss)  for xi in x] for yi in y])
    
    levels = np.arange(-1, 3.001, 1)
    
    norm = pylab.cm.colors.Normalize(vmax=4, vmin=-1)
    cmap = pylab.cm.PRGn
    
    cset1 = pylab.contourf(X, Y, Z, levels,
                            cmap=pylab.cm.get_cmap(cmap,5),
                            norm=norm,
                            hold='on')
    
    cset2 = pylab.contour(X, Y, Z, cset1.levels,
                            colors = 'k',
                            hold='on')
    
    for c in cset2.collections:
        c.set_linestyle('solid')
        
    classesLabels = ["C" + str(i) + " p=" + str(C[i])  for i in range(3)]
    classesColors = ["y", "r", "g"]
    scatterPlots = [pylab.scatter(train[i][:,0], train[i][:,1], c=classesColors[i])  for i in range(3)]
    
    pylab.legend(scatterPlots,
           classesLabels,
           scatterpoints=1,
           loc='lower left',
           ncol=1,
           fontsize=8)
```

```python
plot_regions_and_boundaries2(classify, 0.1)
```

Next, we write the classifier, that will classify each point in the test set and calculate the confounding matrix:

```python
def test_classifier(classify, loss, x_min=0, y_min=-1, x_max=14, y_max=9.3, delta=0.04):
    pylab.clf()
    
    extent = (x_min,x_max,y_min,y_max)
    
    x = np.arange(x_min, x_max, delta)
    y = np.arange(y_min, y_max, delta)
    X, Y = np.meshgrid(x, y)
    Z = np.array([[classify(np.array([xi, yi]), loss)  for xi in x] for yi in y])
    
    levels = np.arange(-1, 3.001, 1)
    
    norm = pylab.cm.colors.Normalize(vmax=4, vmin=-1)
    cmap = pylab.cm.PRGn
    
    cset1 = pylab.contourf(X, Y, Z, levels,
                            cmap=pylab.cm.get_cmap(cmap,5),
                            norm=norm,
                            hold='on')
    
    cset2 = pylab.contour(X, Y, Z, cset1.levels,
                            colors = 'k',
                            hold='on')
    
    for c in cset2.collections:
        c.set_linestyle('solid')

    classified = [[], [], [], []]
    confounding = np.zeros((4, 4))
    for i, pts in enumerate(test):
        for pt in pts:
            cls = classify(pt, loss)
            confounding[i][cls] += 1
            classified[cls].append(pt)
            
    for i in range(4):
        classified[i] = np.array(classified[i])
    
    classesLabels = ["C" + str(i) + " p=" + str(C[i])  for i in range(3)] + ["Rejected loss=" + str(loss)]
    classesColors = ["y", "r", "g", "w"]
    scatterPlots = [pylab.scatter(classified[i][:,0], classified[i][:,1], c=classesColors[i])  for i in range(4)]
            
    pylab.legend(scatterPlots,
           classesLabels,
           scatterpoints=1,
           loc='lower left',
           ncol=1,
           fontsize=8)
    
    return confounding
```

```python
confounding = test_classifier(classify, 0.1)
```

The confounding matrix $C=(c_{ij})_{4\times 4}$ where $c_{ij}$ is the number of examples that should've been classified as belonging to class $C_{i-1}$ but were classified as belonging to class $C_{j-1}$, or were rejected if $j=4$, is given below.

```python
print "Confounding matrix is:"
print confounding
print
print "In percentages:"
print confounding * 100 / np.sum(confounding)
print
```
