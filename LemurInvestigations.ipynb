```python
def h(r):
    return lambda x: r * x * (1 - x)

def funcPow(f, n):
    def g(x):
        for i in range(n):
            x = f(x)
        return x
    return g  
```

```python
import pylab as pl
import numpy as np
%matplotlib inline

def plot(f, xmin, xmax, numpoints=1000):
    x = np.linspace(xmin, xmax, numpoints)
    y = np.array(map(f, x))
    pl.plot(x, y, 'b', x, x, 'g--')
    pl.show()
    
def plotIterations(f, x0, itersFr=0, itersTo=200):
    x = range(itersFr, itersTo)
    y = map(lambda n: funcPow(f, n)(x0), x)
    pl.plot(x, y, 'b.')
    pl.show()

```

```python
for r in 2, 3.1, 4:
    for x0 in 0.25, 0.2345:
        plotIterations(h(r), x0)
```
